```python
# Importing  libraries
import matplotlib.pyplot as plt
import pandas as pd

# Specifing the path to the Excel file
excel_file_path = r'C:\Users\Dell5559\Downloads\Creating a scatter plot of income vs. credit limit.xlsx'

# Reading the Excel file 
data = pd.read_excel(excel_file_path)

# Displaying the first 5  rows of the DataFrame
print(data.head())

# Specifing the columns to be used for x and y axes
x_column = 'Income_Category'
y_column = 'Card_Limit'

# Creating a new figure with 
plt.figure(figsize=(10, 6))

# Creating a scatter plot
plt.scatter(data[x_column], data[y_column], color='blue', alpha=0.5)

# Adding labels and title to the plot
plt.xlabel(x_column)
plt.ylabel(y_column)
plt.title('Scatter Plot of ' + x_column + ' vs. ' + y_column)

# Rotating x-axis labels for better visibility
plt.xticks(rotation=45)

# Adjusting layout for better spacing
plt.tight_layout()

# Displaying the plot
plt.show()
```
